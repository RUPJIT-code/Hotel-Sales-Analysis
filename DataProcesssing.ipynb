```python
import pandas as  pd
data=pd.read_csv('hotel_booking.csv')
df=pd.DataFrame(data=data)
df
```

```python
#Droping the column company because it contains Maximum null values
df.drop(columns='company')
```

```python
# Filling the missing values with 0 which represents "No Agents Used".
df['agent'] = df['agent'].fillna(0).astype(int)
```

```python
# Changing the datatype of the "Children" column to integer
df['children'] = df['children'].replace('', 0).astype(int)
print(df['children'])
```

```python
# Formatting go two letter values to three letter values
df['country'] = df['country'].replace('CN', 'CHN')
```

```python
df.to_csv('hotel_booking_clean.csv',index=False)
```
